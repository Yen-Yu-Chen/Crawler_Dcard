```python
import jieba
jieba.set_dictionary("dict.txt.big")
```

```python
###移除stopword & punctuation
#讀入stopword & punctuation
stopword=[]
punctuation=[]
with open('stopword_chinese.txt', 'r', encoding='utf-8') as file:
    for data1 in file.readlines():
        data1 = data1.strip()
        stopword.append(data1)
with open('punctuation.txt', 'r', encoding='utf-8') as file:
    for data2 in file.readlines():
        data2 = data2.strip()
        punctuation.append(data2)
print(stopword)
print(punctuation)
```

```python
#確認第一則留言是否有移除掉stopword
content = str(content_list[0])
segments = jieba.cut(content, cut_all=False)
#移除stopword & punctuation & "\u3000"(全形的空白符)
remainderWords = list(filter(lambda a: a not in stopword and a not in punctuation and a != "\n",segments))
print(remainderWords)
for k in remainderWords:
    print(k)
```

```python
#移除全部文章的stopword & punctuation & "\u3000"(全形的空白符)
for i in range(103):
    content = str(content_list[i])
    segments = jieba.cut(content, cut_all=False)
#移除stopword & punctuation & "\u3000"(全形的空白符)
    remainderWords = list(filter(lambda a: a not in stopword and a not in punctuation and a != "\n" and a != " " and a != "nan",segments))
    print(remainderWords)
#存成Dcard_comment_232884534_preprocess.txt
```

```python
###TF-IDF & 詞頻分析
#老師講義monpa
import monpa
from sklearn import feature_extraction
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
corpus = []
with open("Dcard_comment_232884534_preprocess.txt", "r", encoding="utf-8") as f:
    for line in f:
        corpus.append(" ".join(monpa.cut(line)))
        
#frequence martix
vectorizer = CountVectorizer()

# a[i][j]: indicate the frequence of term j at document i 
freq_martix = vectorizer.fit_transform(corpus)
transformer = TfidfTransformer()

#計算tf-idf
tfidf = transformer.fit_transform(freq_martix)

#get keyword
words = vectorizer.get_feature_names()

#TF-IDF矩陣
weight = tfidf.toarray()

#確認文章篇數&字詞數
print(tfidf.shape)

#計算每篇文章第一名keyword及其TF-IDF
keyword_index = np.squeeze(np.asarray(np.argmax(weight, axis = 1)))
for i in range(len(weight)):
    print("Keyword:{0}, weighting:{1}".format(words[keyword_index[i]],
                                             weight[i][keyword_index[i]]))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import jieba.analyse
import codecs
from wordcloud import WordCloud
import monpa

jieba.set_dictionary("dict.txt.big")
stopwords = stopword

def generate_wordclound(keywords, stopwords, file_path):
    wc = WordCloud(font_path = "msyh.ttf", background_color = "white",
                   max_words = 2000, stopwords = stopwords)
    wc.generate_from_frequencies(keywords)
    plt.imshow(wc)
    plt.axis("off")
    plt.figure(figsize=(10,6), dpi=100)
    plt.show()
    wc.to_file(file_path)
    
def get_keywords(file_path, topN):
    keywords = {}
    with open(file_path,"r", encoding = 'utf-8') as f:
        corpus = ""
        for line in f:
            corpus= f'{corpus}{" ".join(item for item in monpa.cut(line))}'        
        tags = jieba.analyse.extract_tags(corpus, topK=topN, withWeight=True)
        for tag, weight in tags:
            keywords[tag] = weight
    return keywords

#產生留言的10個關鍵字
keywords = get_keywords("Dcard_comment_232884534_preprocess.txt",10)
print(keywords)
```

```python
#產生文字雲，並儲存
generate_wordclound(keywords, stopwords,"Dcard_comment_232884534_wordclound_0421_monpa_10keyword.jpg")
```

```python
#將結果存成txt檔
file = open('Dcard_comment_232884534_result_0421_monpa_10keyword.txt', 'w') 
file.write("Term"+' '+"G_TF_IDF"+'\n')
for k,v in keywords.items():
        file.write(str(k)+' '+str(v)+'\n')
file.close()
```
